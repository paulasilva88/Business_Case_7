### 1. Os usuários gostam mais de vídeos de qual categoria?
Para se responder a essa pergunta, as seguintes etapas foram seguidas:

```python
# Seleção das colunas que serão utilizadas
dados_likes = dados[['titulo','id_categoria', 'likes', 'visualizacoes']].dropna()
dados_likes.head(5) # perceba que alguns dados estão repetidos
```

```python
# ordenamento pela coluna likes
dados_likes = dados_likes.sort_values(by='likes', ascending=False)

# Remoção das linhas repetidas, de modo que fique apenas os últimos dias, com mais likes
tit_a = 'a'
lista_titulos = []
contador = 0
for titulo in dados_likes['titulo']:
    if titulo in lista_titulos:
        dados_likes = dados_likes.drop(dados_likes.index[contador])
    else:
        tit_a = titulo
        lista_titulos.append(titulo)
    contador +=1

# é preciso executar essa célula mais de uma vez para ela funcionar
```

```python
# visualização do dados_likes, após a remoçao dos dados repetidos
dados_likes.head(5)
```

```python
# Agrupando as categorias por número de likes
categoria_likes = []
for group, frame in dados_likes.groupby('id_categoria'):
    # calculando a soma de likes para a categoria
    soma = np.sum(frame['likes'])
    # inserindo as somas por categoria e suas respectivas categorias em uma lista
    categoria_likes.append([soma, group])
    #print(f'A categoria {group} tem {str(avg)} likes')

# ordenando a lista
categoria_likes.sort(reverse=True)

print(f'Os usuários gostam mais da categoria {int(categoria_likes[0][1])}.')
categoria_likes
```

### 7. Os usuários comentam em qual categoria mais? 
Requisito analisado
- Categoria com mais comentários

```python
# Seleção das colunas que serão utilizadas
comentario = dados[['titulo','id_categoria', 'qtd_comentarios']]
comentario.head(1)
```

```python
# ordenamento pela coluna de comentarios
comentario = comentario.sort_values(by=['qtd_comentarios'], ascending=[False]).head(1000) ## Amostra de dados para análise

# Remoção das linhas repetidas
tit_a = 'a'
lista_v = []
contador = 0
for titulo in comentario['titulo']:
    if titulo in lista_v :
        comentario = comentario.drop(comentario.index[contador])
    else:
        tit_a = titulo
        lista_v.append(titulo)
    contador +=1
comentario.head()
```

```python
# Agrupando as categorias por número de comentários
coment = []
for group, frame in comentario.groupby('id_categoria'):
    # calculando a soma de comentarios por categoria
    soma_coment = np.sum(frame['qtd_comentarios'])
    # somando as somas a lista com a soma por categoria e suas respectivas categorias
    coment.append([soma_coment, group])
    #print(f'A categoria {group} tem {str(avg)} likes')
coment.sort(reverse=True)
print(f'Os usuários comentam mais nos vídeos da categoria {int(coment[0][1])}.')
```
